```python
import numpy as np
import scipy.integrate as integrate
import matplotlib.pylab as p
```

```python
def dMS_dt (pops,t=0):  
  M = pops [0]
  S = pops [1]
  Mmax=1.35
  S=0.000022
  Ks=3.76
  μο=0.01
  qo=0.01
  Kp=10
  p=0.0000002
  Ksa=0.87
  Ksb=0.87
  X=0.0000002
  Yg=0.4036
  Yp=0.4036
  dXdt=(μο*Kp/(Kp+p))*(S/(Ksa+S))*X
  dPdt=(qo*Kp/(Kp+p))*(S/(Ksb+S))*X
  dSdt=dXdt/Yg+dPdt/Yp+M*X
  dMdS=Mmax*(dSdt/(Ks+dSdt))
  y= np.array([dMdS,dSdt])
  dMdS
  return np.array([dMdS,dSdt])
  np
t = np.linspace (0, 15, 1000)
M0=0
S0=0
MS0=np.array([M0,S0])
pops, infodict=integrate.odeint(dMS_dt,MS0,t,full_output=True)
pops
type(infodict)
infodict.keys()
infodict['message']
p.plot (t, pops[:,0], 'g-', label='growth rate (1/hr)') # Plot
p.plot (t, pops[:,1], 'b-' , label='concentration of the limiting substrate (mg/l)')
p.grid()
p.legend(loc='best')
p.xlabel ('time')
p.ylabel 
p.title ('Monod Equation')
p.show()# To display the figure
```
