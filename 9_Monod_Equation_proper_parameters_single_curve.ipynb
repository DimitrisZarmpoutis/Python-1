```python
import numpy as np
import scipy.integrate as integrate
import matplotlib.pylab as p
```

```python
def dMdS (pops,t=0):  
  M = pops [0]
  Mmax=1.35
  S=0.000022
  Ks=3.76
  μο=0.01
  qo=0.01
  Kp=10
  P=0.0000002
  Ksa=0.87
  Ksb=0.87
  X=0.0000002
  Yg=0.4036
  Yp=0.4036
  dXdt=(μο*Kp/(Kp+P))*(S/(Ksa+S))*X
  dPdt=(qo*Kp/(Kp+P))*(S/(Ksb+S))*X
  dSdt=dXdt/Yg+dPdt/Yp+M*X
  dMdS=Mmax*(dSdt/(Ks+dSdt))
  y= np.array([dMdS,dSdt])
  dMdS
  return np.array([dMdS,dSdt])
  np
S = np.linspace (0, 15, 1000)
M0=0
S0=0
MS0=np.array([M0,S0])
pops, infodict=integrate.odeint(dMdS,MS0,S,full_output=True)
pops
type(infodict)
infodict.keys()
infodict['message']
p.plot (t, pops[:,0], 'g-', label='growth rate') # Plot
p.grid()
p.legend(loc='best')
p.xlabel ('concentration of the limiting substrate')
p.ylabel ('growth rate')
p.title ('Monod Equation')
p.show()# To display the figure
```
