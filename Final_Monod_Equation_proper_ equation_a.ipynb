```python
import numpy as np
import scipy.integrate as integrate
from scipy.integrate import odeint
import matplotlib.pylab as plt
```

```python
def f(r,t):
  Mmax=1.35
  S=0.000022
  Ks=3.76
  Yg=0.4036
  M = 5
  S = 0.05
  dMdt=(S/(Ks+S))*M*Mmax
  dSdt=-M*S/Yg
  return [dMdt,dSdt]
  np
t = np.linspace (0, 15, 1000)
r0=[0.05,10]
r=odeint(f,r0,t)
plt.plot (t, r[:,0], 'g-', label='growth rate') # Plot
plt.plot (t, r[:,1], 'b-', label='concentration of the limiting substrate')
plt.grid()
plt.legend(loc='best')
plt.xlabel ('concentration of the limiting substrate (mg/l)')
plt.ylabel ('growth rate (1/hr)')
plt.title ('Monod Equation')
plt.show()# To display the figure
```
