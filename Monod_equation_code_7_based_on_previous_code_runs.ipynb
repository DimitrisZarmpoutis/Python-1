```python
import numpy as np
import scipy.integrate as integrate
import matplotlib.pylab as p
```

```python
def dMS_dt (pops,t=0):  
  M = pops [0]
  S = pops [1]
  Mmax=15
  S=0.1
  Ks=1.5
  μο=1
  qo=1
  Kp=1
  p=1
  Ksa=1
  Ksb=1
  X=1
  Yg=1
  Yp=1
  dMdt=Mmax*(S/(Ks+S))
  dXdt=(μο*Kp/(Kp+p))*(S/(Ksa+S))*X
  dPdt=(qo*Kp/(Kp+p))*(S/(Ksb+S))*X
  dSdt=dXdt/Yg+dPdt/Yp+M*X
  y= np.array([dMdt,dSdt])
  dMdt
  return np.array([dMdt,dSdt])
  np
t = np.linspace (0, 15, 1000)
M0=0
S0=0
MS0=np.array([M0,S0])
pops, infodict=integrate.odeint(dMS_dt,MS0,t,full_output=True)
pops
type(infodict)
infodict.keys()
infodict['message']
p.plot (t, pops[:,0], 'g-', label='growth rate') # Plot
p.plot (t, pops[:,1], 'b-' , label='concentration of the limiting substrate')
p.grid()
p.legend(loc='best')
p.xlabel ('time')
p.ylabel ('growth rate')
p.title ('growth rate of a microorganims')
p.show()# To display the figure
```
